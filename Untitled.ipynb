```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

we are using pil i.e Pillow is a modern, maintained fork of the original Python Imaging Library (PIL) . as Pillow is for simpler tasks that require basic image manipulation.

we will use opencv when we work in more advance problem, and in deepfake using video as OpenCV is used for advanced computer vision and real-time processing needs.

```python
real_images_path = 'dataset/Train/Real_trial'
fake_images_path = 'dataset/Train/fake_trial'
```

```python
# Function to load images from a folder
def load_images_from_folder(folder, label):
    images = []   # defining empty list which will store 3d arrays of images
    labels = []   # defing empty list which will store the lable i.e 0 for real and 1 for fake, this lable will be provided in argument along with folder of real or fake
    
    for filename in os.listdir(folder):    # os.listdir(folder) returns a list of filenames in folder one by one
        
        img_path = os.path.join(folder, filename)   #platform-independent way of creating file paths by joing folder path  and the filename 
        if filename.endswith('.jpg') or filename.endswith('.png'):    # if its an image
            img = Image.open(img_path)
            img = img.resize((128, 128))  # Resize image to a manageable size
            img = np.array(img)
            images.append(img)
            labels.append(label)  # 0 for real, 1 for fake
    return images, labels
```

```python
# Load real and fake images
real_images, real_labels = load_images_from_folder(real_images_path, 0)
fake_images, fake_labels = load_images_from_folder(fake_images_path, 1)

# Combine both datasets to one with real and fake together along with corresponding lable so we can enter it in our model
images = np.array(real_images + fake_images)
labels = np.array(real_labels + fake_labels)

print(f"Loaded {len(images)} images")    # this will show the no of  total images we loaded
```

### Visualize some images:

```python
# Display some sample images
def display_images(images, labels, n_samples=5):   # n_sample is no of smaple( default 5)
    plt.figure(figsize=(15, 8))                    # plot size
    
    for i in range(n_samples):
        plt.subplot(1, n_samples, i+1)
        plt.imshow(images[i])
        plt.title('Fake' if labels[i] == 1 else 'Real')
        plt.axis('off')
    plt.show()

display_images(images, labels, 5)
```

```python
import seaborn as sns

# Visualize the class distribution (real vs fake)
sns.countplot(x=labels)      # x=labels parameter indicates that the function should count the occurrences of each unique value in the labels array
plt.title("Class distribution: Real vs Fake")
plt.xticks([0, 1], ['Real', 'Fake'])    # [0, 1] specifies the positions of the ticks.  ['Real', 'Fake'] specifies the labels 
plt.show()
```

### Preprocessing


```python
# Flatten the images to 1D (since Logistic Regression requires 1D input)
images_flattened = images.reshape(images.shape[0], -1)

# Normalize the pixel values (between 0 and 1)
images_flattened = images_flattened / 255.0
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images_flattened, labels, test_size=0.2, random_state=42)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images_flattened, labels, test_size=0.2, random_state=42)  # i.e 80% train 20% test
```

```python

# Scale the features (Standardization)
scaler = StandardScaler()        #Standardization puts all the features on the same scale, which allows the model to treat them equally and perform better.
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

###  Train a Logistic Regression Model:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Initialize Logistic Regression model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)

cm = confusion_matrix(y_test, y_pred)  # confusion matrix (cm), which is a table that is used to evaluate the performance of the classification model. 
```

```python
print(f"Accuracy: {accuracy*100:.2f}%")
print("Confusion Matrix:")
print(cm)          # diaonal tell the no of which the answer is right
```

```python
# Plot confusion matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import classification_report

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python

# Initialize Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy*100:.2f}%")
print("Confusion Matrix:")
print(cm)

# Plot confusion matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.title("Confusion Matrix")
plt.show()
```
